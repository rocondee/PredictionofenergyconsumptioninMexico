```python
import io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import load_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
new_model = load_model('model_RNN.h5')
new_model.summary()
```

```python
df = pd.read_csv('Dataset.csv')
df.dropna(inplace=True) 
df_train, df_test = train_test_split(df, train_size=0.6, test_size=0.4, shuffle=False)

X = df_train.loc[:, :].values
min_max_scaler = MinMaxScaler()  
X_train = min_max_scaler.fit_transform(X) 
X_test = min_max_scaler.transform(df_test.loc[:, :].values) 

def build_dataset(X, Y, hist_size):
  '''
  Params
    X: data matrix [n_time_steps, n_X_feats]
    Y: label matrix [n_time_steps, n_Y_feats]
    hist_size: integer indicating the number of time steps in each sample of x
  Returns
    x: tensor of input data [n_samples, n_timesteps, n_X_features]
    y: tensor of output data [n_samples, n_Y_features]
  '''
  n_samples = Y.shape[0] - hist_size
  x = np.zeros((n_samples, hist_size, X.shape[1])) 
  y = np.zeros((n_samples))

  for ind in range(n_samples):         
    x[ind] = X[ind : ind + hist_size]  
    y[ind] = Y[ind + hist_size]        
  return x, y

time_steps = 2
x_train, y_train = build_dataset(X_train[:, 0:5], X_train[:, 6], time_steps)
x_test,  y_test  = build_dataset(X_test[:, 0:5],  X_test[:, 6],  time_steps)

y_train_hat = new_model.predict(x_train)
y_test_hat = new_model.predict(x_test)
```

```python
X = df_train.loc[:, :].values

scaler = MinMaxScaler(feature_range=(min(X[:,6]), max(X[:,6])))

Y1=pd.DataFrame(y_train)
Y2=pd.DataFrame(y_train_hat)
Y3=pd.DataFrame(y_test)
Y4=pd.DataFrame(y_test_hat)
scaler = scaler.fit(Y1)
Y1c= scaler.transform(Y1)
Y2c= scaler.transform(Y2)
Y3c= scaler.transform(Y3)
Y4c= scaler.transform(Y4)

#Graphic
YReal = np.concatenate((Y1c, Y3c))
YHat = np.concatenate((Y2c, Y4c))
año1= df.iloc[2:33,0]
año2= df.iloc[36:56,0]
y=YReal[0:31,0]
y2= YReal[32:52,0]
y3=YHat[0:31,0]
y4= YHat[32:52,0]
fig, ax = plt.subplots(figsize=(12,8 )) 
aoFull=np.linspace(1,8,57)
yy=np.zeros(57)
x = yy+1999.5
plt.plot(x,aoFull, 'k--')
plt.plot(año1, y, "b^", label='Real data')
plt.plot(año2, y2, "b^")
plt.plot(año1, y3, "k-", label = 'Simulated data')
plt.plot(año2, y4, "k--", label = 'Forecast data')
L = ax.legend()
plt.setp(L.texts, family='Times New Roman', fontsize=20)
plt.xlabel("Year", fontsize=20, color="k", 
         name="Times New Roman")
plt.ylabel("Energy consumption [EJ]", fontsize=20, color="k", 
         name="Times New Roman")
ax.spines[ "top"].set_color ( "None")
ax.spines["right"].set_color ("None")
plt.xticks(fontsize=20, name='Times New Roman')
plt.yticks(fontsize=20, name='Times New Roman')
plt.show()
```
